3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

```
def check_number():
    while type:
        number=input("Input power of Newton's binomial:")
        try:
            number=int(number)
            if type(number)==int:
                if number>=0:
                    return number
                else:
                    print("Number must be more or equal 0")
        except ValueError:
            print( "Value Error")



def binom(limit):
    a=[1]
    print(*a)
    i=0
    while i<limit:
        b=[1]
        b+=[a[k]+a[k+1] for k in range(len(a)-1)]+[1]
        a=b
        yield a
        i+=1

for i in binom(check_number()):
    print(*i)
```
